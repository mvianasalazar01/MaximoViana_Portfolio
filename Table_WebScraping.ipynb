```python
import requests 
from bs4 import BeautifulSoup

url = 'https://www.worldometers.info/world-population/population-by-country/'
requests.get(url)
```

```python
page = requests.get(url) 
soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find('table','table table-striped table-bordered')
```

```python
world_columns = table.find_all('th')
```

```python
all_columns = [columns.text for columns in world_columns]
print(all_columns)
```

```python
import pandas as pd
df = pd.DataFrame(columns = all_columns) 
```

```python
world_rows = soup.find_all('tr')
for row in world_rows:
    row_data = row.find_all('td')
    if row_data:  
        table_row = [data.text.strip() for data in row_data]
        df.loc[len(df)] = table_row 
df
```
